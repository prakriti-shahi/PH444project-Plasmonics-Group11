# Analytical calculation of plasmonic resonances in metal nanoparticles: A simple guide

#Introduction

Plasmonics is a field of physics that lies at the intersection of nanotechnology and photonics. Plasmonics involves the creation, study, and manipulation of signals embedded in optical-frequency oscillations of surface electrons along metal-dielectric interfaces. In plasmonics, optical-frequency signals are confined to subwavelength-size volumes, thus providing an interface optical electromagnetic fields and nanoscopic devices. These interactions lead to charge oscillations called plasmons. When the frequency of the EM field matches the resonant frequency of the metal nanoparticle, the plasmon is excited. The electron is free to move within the boundary of the particle surface, but is confined to the surface and hence this effect is called a localized surface plasmon (LSP) or plasmonic resonance. This finds applications in fields such as nanophysics, nanochemistry and nanobiology. While the quantum mechanical nature of nanoparticles makes their properties counter-intuitive, luckily the prominent plasmonic effects can be understood in a classical framework. We will begin with a historical perspective. Then, we will need a framework to describe three distinct features: the metallic characteristic, the size, and the shape of the particle. We start with the Drude model and then look at the quasi-static approximation which will let us take into account the nanoscale size of the particles. Then, the generalization of the Clausius–Mossotti relation will let us consider a particle embedded in a dielectric medium and the effect of its shape. We combine these to build a classical model for the plasmonic resonance in metal nanoparticles. We will then deal with some of the limitations of the quasi-static approximation by introducing final corrections to the model, to extend its validity as far as possible. As our assumptions are all based on classical electromagnetism, the final model will be classical too.

# Brief History

Although nanotechnology hadn't emerged as a scientific field until a few decades ago, lead-based quantum dots had been used by ancient Greeks and Romans, and copper nanoparticles were employed in red opaque glass production in Egypt and Mesopotamia. Michael Faraday iperformed the first scientific experiments on the optical properties of nanoparticles, focusing on gold colloids in the 1850s. He was puzzled by the ruby red color of the solutions he synthesized, far removed from the aureate color of bulk gold. A satisfying classical explanation of this phenomenon came only in 1904, when Maxwell Garnett combined the new Drude theory of metals with Lord Rayleigh’s description of electromagnetic properties of small spheres. Shortly after, in 1912, Richard Gans successfully extended the description of optical phenomena to oblate and prolate spheroids. However, these theories were all purely based on the optical properties of bulk metals, and it was not until 1970 that they were modified by Uwe Kreibig and Peter Zacharias to take their nanoscale size into account. It became clear that size, shape, arrangement, medium, and temperature all have a crucial role in controlling the intensity and frequency of plasmonic resonances. Let us start by examining the object under study, the metal nanoparticle, and its defining features.

## A. Drude model for metals
The electric properties of materials are characterized by
 the dielectric constant , which relates the electric displacement $D⃗$ the electric field $E⃗$ through the equation  $D⃗$ = $\epsilon_o$$E⃗$ + $P⃗$ = $\epsilon_o$$ϵ$$E⃗$  . When the applied field is sinusoidal, the dielec
tric constant is a complex number; the imaginary part represents the out-of-phase response of the material. \\
 \\
 We use the lorentz oscillator model which assumes that the electron undergoes damped oscillations under forcing( which is the external field ) and the damping is caused by collisions with other particles and the nuclear potential is assumed to be spring like. \\
 $ m\ddot{x} = F_E - F_{el} - F_{\gamma} \implies m\ddot{x} + m\gamma\dot{x} + m\omega_0^2 x = e E(t) \tag{1}$ \\
  $γ$ is the damping constant and $\omega_0$ is the natural frequency \\
   $γ$
 is given by Matthiessen’s rule for different independent collision events, each possessing a
 time constant $τ_j$ - $ \gamma = \sum_{j} \frac{1}{\tau_j}. \tag{2}$
 We take $E(t) = E_0e^{iωt}$ and use the following ansatz -
 $ x(t) = x_0 e^{i\omega t}, \quad \dot{x}(t) = x_0 (i\omega) e^{i\omega t}, \quad \ddot{x}(t) = -x_0 (\omega^2) e^{i\omega t}. \tag{3}$
 Putting the following ansatz back in the diffrential equation we get - \\
 $x_0 = \frac{e E_0}{m} \frac{1}{\omega_0^2 - \omega^2 + i\gamma\omega}. \tag{4}$
 $x(t) = -\frac{e}{m} \frac{1}{\omega_0^2 - \omega^2 + i\gamma\omega} E(t) \tag{5}$
 We can calculate dipole moment -$ p(t) = -e x(t) = -\frac{e^2}{m( \omega_0^2 - \omega^2 + i \gamma \omega)} E(t) \tag{6}$ \\
$ P(t) = -N p(t) = \frac{Ne^2}{m} \frac{1}{\omega_0^2 - \omega^2 + i \gamma \omega} E(t) \tag{7}$
Using  $D⃗$ = $\epsilon_o$$E⃗$ + $P⃗$ = $\epsilon_o$$ϵ$$E⃗$ we get - \\
 $\epsilon = 1 + \frac{Ne^2}{\varepsilon_0 m} \frac{1}{\omega_0^2 - \omega^2 + i \gamma \omega} \tag{8}$ \\
  Take $ ϵ = ϵ_1 - iϵ_2 $ \\
 $\epsilon_1 = 1 + \frac{Ne^2}{\varepsilon_0 m} \frac{\omega_0^2 - \omega^2}{(\omega_0^2 - \omega^2)^2 + (\gamma \omega)^2}\tag{9} $ \\
$\epsilon_2 = \frac{Ne^2}{\varepsilon_0 m} \frac{\gamma \omega}{(\omega_0^2 - \omega^2)^2 + (\gamma \omega)^2}\tag{10}$ \\
In the Drude model we assume that the electrons are free and do not experience any force from the nucleus so $\omega_0$ = 0 . \\
$\epsilon_m = 1 + \frac{Ne^2}{\varepsilon_0 m} \frac{1}{-\omega^2 + i \gamma \omega} = 1 - \frac{\omega_p^2}{\omega^2 - i \gamma \omega} \tag{11}$
where $ω_p = \sqrt{\frac{Ne^2}{ϵ_0m}}$ which is the plasma frequency \\
A clearer formulation splits $ϵ_m$ in
 its real and imaginary components: \\
 $\epsilon_m = 1 - \frac{\omega_p^2}{\omega^2 + \gamma^2} - i \frac{\omega_p^2 \gamma}{\omega^3 + \gamma^2 \omega} \tag{12}$ \\
  Furthermore, an extension to the Drude model has
 been formulated to account for the positive background of
 ions in the constant term $ϵ_{m,∞}$, referred to as the core
 polarization. \\
  $\epsilon_m = ϵ_{m,∞} - \frac{\omega_p^2}{\omega^2 + \gamma^2} - i \frac{\omega_p^2 \gamma}{\omega^3 + \gamma^2 \omega} \tag{13}$

## B. Quasi-Static Approximation

To find the polarization of a metal nano-particle by the electric field, a good start to such a treatment would be by taking the Quasi Static Approximation (QSA). In this scheme, if the nano-particle is much smaller than the wavelength of the applied electric field, then all the electrons inside the nanoparticle feel the same electric field and oscillate in phase.

If $d$ is the size of the nanoparticle, $λ$ the wavelength, $k$ the wave-vector, $ω$ the angular frequency of the incoming field $\vec{E_o}$, $\vec{r}$ the relative position of an arbitrary point with respect to the center of the nanoparticle, and $t$ the time, we get

$\vec{E}^0(\vec{r}, t) = \vec{E}^0 e^{i(\omega t - \vec{k} \cdot \vec{r})} \approx \vec{E}^0 e^{i\omega t}. \tag{14}$

If $d<<λ$ and hence $\vec{k}.\vec{r} ≈0$ (since $2πd/λ$ tends to 0 since d is much smaller).

With the electric field not depending upon $\vec{r}$, the polarization is uniform, and the effect of the quasi-static approximation on the total dipole moment of the nanoparticle $\vec{p}_{np}$ is very conveniently translated to (because of polarization just being linear and uniform)

$\vec{p}_{np} = VP \tag{15}$

where $V$ is the volume of the nanoparticle and $\vec{P}$ is the polarization density. The quasi-static approximation lets us drop the dependence on the position $\vec{r}$.

In this treatment, we assume that the skin depth (the distance the field has to penetrate the metal for it's amplitude to decrease by a factor of 1/e) is larger than the size of the nano-particle, hence allowing us to assume that the field can penetrate the nanoparticle completely and the uniformity of the polarization can still be assumed





## C. Generalization of the Clausius–Mossotti relation for ellipsoids

To find the polarizability of the nano-particle with the external force applied, we model the nano-particle as an isolated ellipsoid having a complex dielectric function $ϵ_m$ and semiaxes, respectively, $a_x$, $a_y$, and $a_z$, immersed in a dielectric material having a dielectric constant $ϵ_h$ (h subscript stands for the host medium surrounding the nano-particle)

We now need to find \\

$\vec{P}_{np} = \varepsilon_0 \hat{\alpha} \vec{E}_{loc} = \varepsilon_0 \varepsilon_h \hat{\alpha} \vec{E}^{0}
\tag{16}$ \\

(if the dielectric medium is assumed to be linear, homogeneous and isotropic, the local field is proportional to the applied electric field $\hat{E}_{loc} = \varepsilon_h \vec{E}^{0}$) \\
The total electric field will depend upon the external electric field and also the polarization caused due to movement of charges . \\

The surface charges lead to a field  inside the dielectric, sometimes called the depolarization field (depolarization because, in the static limit, the field produced by the displaced charges acts to counter the applied field). \\
$E_{net} = E_{ext} + E_{pol}$ \\
 \\

If we take a gaussian pillbox as given in the diagram above only the depolarizing field will contribute to the integral as flux due to external field is zero. \\
$E_{pol} = -\frac{P}{ϵ_o}$ \\
$E_{net} = E_{ext} - \frac{P}{ϵ_o}$ \\
Screenshot 2024-04-06 024544.png \\


   
But in this case the nano particle is embedded in another medium($ϵ_h$) where the induced charge in the nanoparticle leads to more induced charge in the medium at the interface which leads to reduction in polarization. \\
The depolarizing field also depends on the shape of the nanoparticle which was not seen in the infinite slab case . \\
 $E_{\text{loc},j} = E_{j}^{0} - \frac{L_j P_{j}}{\varepsilon_0\varepsilon_h}, \tag{17} $ \\

 $j=(x,y,z)$ \\
 $L_j$ is the depolarizing factor which depends on the shape of the nanoparticle. \\
 For a sphere $L_j = \frac{1}{3}$ and for an  elliptical nanoparticle $L_j$ is given below
$ L_j = \frac{a_x a_y a_z}{2} \int_{0}^{\infty} \frac{dq}{(q + a_j^2)\sqrt{\prod_{\eta=x,y,z} (q + a_{\eta}^2)}}, \tag{18} $ \\

 $ P_j = P_{m,j} - P_{h,j} = \varepsilon_0 \varepsilon_m E_{\text{loc},j} - \varepsilon_0 \varepsilon_h E_{\text{loc},j} = \varepsilon_0 (\varepsilon_m - \varepsilon_h) E_{\text{loc},j} \tag{19} $ \\

 Combining equations $(17)$  and $(19)$ we get \\
 $P_j = \varepsilon_0 (\varepsilon_m - \varepsilon_h) \left( E_{j}^{0} - \frac{L_j P_{j}}{\varepsilon_0\varepsilon_h} \right). \tag{20}$ \\

 Solving for $P_j$ , \\
 $P_j = \frac{\varepsilon_0 \varepsilon_h (\varepsilon_m - \varepsilon_h)}{\varepsilon_h + L_j(\varepsilon_m - \varepsilon_h)} E_j^0. \tag{21}$ \\
Now using the quasistatic approximation given in equation $(5)$ we get \\

 $p_{np,j} = \frac{\varepsilon_0 \varepsilon_h V(\varepsilon_m - \varepsilon_h)}{\varepsilon_h + L_j(\varepsilon_m - \varepsilon_h)} .E_J^0 \tag{22}$
 Which implies that \\
 $\alpha_j = \frac{4 \pi a_x a_y a_z}{3} \frac{\varepsilon_m - \varepsilon_h}{\varepsilon_h + L_j(\varepsilon_m - \varepsilon_h)}. \tag{23}$ \\
 We can see that $α_{ij}$ is a diagonal matrix which means that if the incoming field is along one of the axes of the ellipsoid the polarization is parallel to it. Note how it being a matrix also makes the system anisotropic. \\
Calculating the depolarization factor integral is a complicated task. We can generally use a property of $L_j$, which is \\

$\sum_{j=x,y,z} L_j = 1. \tag{24}$

Using such this normalisation property, finding values for $L_j$ for objects with symmetries or objects with one dimension being too large can be a trivial task. The table below summarizes a few results.

With $L_j=1/3$ for spherical particles, we can finally write the Clausius–Mossotti relation - \\
$\alpha_j = 3v\frac{\varepsilon_m - 1}{\varepsilon_m + 2}, \quad j = x, y, z. \tag{25}$ \\
It relates the microscopic property of polarizability (a measure of how much an individual molecule will be distorted by an electric field) to the macroscopic property of the dielectric constant (which characterizes how the material as a whole responds to an electric field).


## D. Plasmon Resonance

From Drude's theory, we know that \\

$\epsilon_m(\omega) = \epsilon_{m,\infty} - \frac{\omega_p^2}{\omega^2 + \gamma^2} - i \frac{\omega_p^2 \gamma}{\omega^3 + \gamma^2 \omega}. \tag{13}$

From our calculation of the polarizability matrix for a general ellipsoidal particle, we also know that \\

$\alpha_j = \frac{4 \pi a_x a_y a_z}{3} \frac{\varepsilon_m - \varepsilon_h}{\varepsilon_h + L_j(\varepsilon_m - \varepsilon_h)}. \tag{23}$ \\

For a metal nanoparticle, we can plug in equation $(3)$ in $(13)$. Assuming gold-like Drude particles, \\

 $ϵ_{m,∞} = 7.926\ , ω_P = 1.23 *10^{16} \ [rad/s] \ , γ = 3.8*10^{13}\ [rad/s]\tag{25}$ \\

 We can find the relation between $|α|$ and the wavelength.

```python
# CODE TO PLOT POLARIZABILITY VERSUS WAVELENGTH OF THE APPLIED ELECTRIC FIELD

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import math

colors = ['red', 'blue']


# GOLD LIKE DRUDE PARAMETERS
omega_p = 1.23e16
gamma = 3.8e13
epsilon_m_inf = 7.93
c = 3e8
epsilon_h_vac = 1
epsilon_h_water = 1.77

for lambda_w in np.linspace(400, 600, num = 301, endpoint = True):

    omega = (2*math.pi*c)/(lambda_w*1e-9)
    epsilon_m = complex(epsilon_m_inf - ((omega_p**2)/(omega**2 + gamma**2)), -((omega_p**2 * gamma) / (omega**3 + (gamma**2 * omega))) )
    #print(epsilon_m)
    alpha_j = (3*(epsilon_m - epsilon_h_vac))/(epsilon_m+(2*epsilon_h_vac))

    #print(alpha_j)
    plt.plot(lambda_w, abs(alpha_j), marker = 'o', markersize = 2, color = 'red')

    alpha_j = (3*(epsilon_m - epsilon_h_water))/(epsilon_m+(2*epsilon_h_water))

    #print(alpha_j)
    plt.plot(lambda_w, abs(alpha_j), marker = 'o', markersize = 2, color = 'blue')

plt.xlabel('Wavelength (nm)')
plt.ylabel('|α|/v [a.u.]')
red_patch = mpatches.Patch(color='red', label='$\epsilon_h$ = 1')
blue_patch = mpatches.Patch(color='blue', label='$\epsilon_h$ = 1.77')
plt.legend(handles=[red_patch, blue_patch])
```

The functional form for $|α|$ is a sharp peak. At the angular frequency where the polarizability is maximal, the electrons oscillate with a higher amplitude. Given such a sharp feature in the frequency, we refer to the peak position as the resonance frequency. \\
We can find an approximate value of $ϵ_m$ by making the denominator approach 0. Since $ϵ_h$ is real, we get \\

$\text{Re}[\epsilon_m(\omega_{\text{res}})] = -\frac{1 - L_j}{L_j} \epsilon_h, \quad \text{Im}[\epsilon_m(\omega_{\text{res}})] \approx 0 \tag{26}$ \\

(Notice how the Imaginary component of $\epsilon_m$, which gives us the damping of oscillations is approximately 0) By plugging in the approximate values of equation $(16)$ into the Drude equation $(3)$, we get \\
$\omega_{\text{res}} = \sqrt{\frac{\omega_p^2}{\frac{1 - L_j}{L_j} \epsilon_{m,\infty} + \epsilon_{h} - \gamma^2}} \approx \frac{\omega_p}{\sqrt{\frac{1 - L_j}{L_j} \epsilon_{m,\infty} + \epsilon_{h}}}. \tag{27}$ \\

These equations clearly show that resonance frequency depends on the geometrical features of the nanoparticle and on the dielectric medium in which it is immersed. The dependence of the resonance shift due to changes in $ϵ_h$ enables a multitude of sensing applications. In the host dielectric environment, variations due to chemical, physical or biological activity result in fact in shifts in plasmonic peaks.




For the special case of spherical particles, you get the resonance frequency of the metal nano-particle to be \\
$\text{Re}[\epsilon_m(\omega_{\text{res}})] = -2\epsilon_h, \quad \text{Im}[\epsilon_m(\omega_{\text{res}})] \approx 0,\tag{28}$ \\
$\omega_{\text{res}} \approx \frac{\omega_p}{\sqrt{\epsilon_{m,\infty} + 2\epsilon_h}}. \tag{29}$ \\

Note that for resonance to exist, $ \text{Im}[\epsilon_m(\omega_{\text{res}})] \approx 0$, just like our relatively simple models show. This is to keep damping at a minimum. For this reason, the most significant plasmonic materials are also the most conductive: silver, gold, copper, and aluminum.

One promising aspect of Plasmonic Resonance. Outside our nanoparticle, we can assume the build up of an electric dipole inside the particle. This can give us the local field outside the nanoparticle which takes the shape of a classical dipole field \\

$\vec{E}(\vec{r}) = \vec{E}^0 + \frac{1}{4\pi\varepsilon_0\varepsilon_h} \frac{3(\vec{r} \cdot \vec{P}_{np})\vec{r} - r^2\vec{P}_{np}}{r^5}, \tag{30}$

As can be intuited, resonances in a will reflect on $p_{np}$ and subsequently on $\vec{E}(\vec{r})$. This enhancement due to plasmonic resonance can make the local electric field tens or hundreds of times higher than the incoming field. The shape of the particle can be engineered to create radiation enhancements that are wavelength, polarization, and direction-dependent, just like in classical antennas

#BEYOND QUASISTATIC APPROXIMATION
 The quasi-static approximation describes nanoscale processes fairly well, such as the absorption of certain colors or
 the local field enhancement in the proximity of metal nano
particles. However, it fails to predict other effects related to
 particle size: in the quasi-static approximation, particle size
 is irrelevant, as long as it is smaller than the incoming wave
length.
 The first correction we will introduce takes into account
 the collisions of the electrons with the nanoparticle bound
aries, which have implicitly been neglected until now. Then,
the second correction will take us closer to the multipolar   regime by involving retardation effects.
##  A. Surface Damping
 Till now we have only considered damping due to collisions of electrons with other electrons , nuclei and phonons .The electrons also collide with the particle boundary , more so if the particle size is much smaller than the mean free path of the electron. In this case the electron will collide with the particle boundary more than any other object.

 The free electrons move wih fermi velocity $\\v_f$. \\

 The time between two collisions with the boundary is $τ_{boundary}$. \\

  $τ_{boundary}$ = $\frac{λ_{boundary}}{ \\v_f}\tag{31}$

$λ_{boundary}$ is the distance travelled between successive coliisions.

$γ$ = $\sum_{j}1/τ_{j}\tag{32}$ \\

The above equation is also known as the Mathiessens's Rule. \\
$γ$ = $γ_{bulk}$ +$\frac{1}{τ_{boundary}}$ = $\frac{v_f}{λ_{MFP}}$ + $\frac{v_f}{λ_{boundary}}\tag{33}$ \\
$λ_{MFP}$ is the mean free path . \\


Lets say we have a spherical nanoparticle of radius $r$ and the electron  first collides at point $O(x,y,z)$ and then finishes at $O(x',y',z')$ and it travels a distance of $d[O,O']$ . \\
$ x^2 + y^2 + z^2 = r^2$ and  



Untitled_page-0001.jpg


We can assume $O(x,y,z)$ to be $(0,0,r)$ without any loss in generality and all other points $O'$ have the same probability of getting hit . \\
$d[O,O'] = \sqrt{x'^2 + y'^2 + (z'-r)^2}  = \sqrt{2}r\sqrt{1-\frac{z'}{r}}\tag{34}$  \\
$λ_{boundary}$ is basically the average value of $d[O,O']$ . \\
$λ_{boundary} = \frac{\int_{0}^{2π}\int_{-r}^{r}d[O,O']dz'dθ}{\int_{0}^{2π}\int_{-r}^{r}dz'dθ}\tag{35}$ \\

Solving the trivial integral above we get value of $λ_{boundary}$ to be $\frac{4r}{3}$. \\
$ϵ_{m}= ϵ_{m,∞} - \frac{ω_p^2}{ω^2 + γ^2} - i\frac{ω_p^2γ}{ω^3 + γ^2ω}=ϵ_1-iϵ_2\tag{36}$ \\
$ω >> γ $ in realistic cases \\
$ϵ_1 ≈ ϵ_{m,∞} - \frac{ω_p^2}{ω^2}\tag{37}$ and $ϵ_2 ≈ \frac{ω_p^2γ}{ω^3}\tag{38}$ \\
As we can see that the real part does not change due to damping , only the imaginary part changes. \\
$ϵ_2 ≈ \frac{ω_p^2}{ω^3}(γ_{bulk} + \frac{3v_f}{4r}) = ϵ_{2_{bulk}} + A\frac{v_f}{r}\tag{39}$  \\
     $A=  \frac{4ω_p^2}{3ω^3}\tag{40}$   \\
So when we include damping in the picture , the imaginary part of the dielectric constant($ϵ_2$) is not negligible anymore which was crucial in our condition for resonance earlier . This leads to a diminished resonance  peak in α .









```python
# CODE FOR PLOTTING POLARIZABILITY VS WAVELENGTH WITH SURFACE DAMPING
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

colors = ['red', 'blue', 'green', 'cyan', 'purple']

# Constants
ω_p = 1.23e16  # rad/s
γ = 3.8e13  # rad/s
ε_m_inf = 7.93
ε_h = 1.77
v_f = 1.39e6  # m/s
#r = 100e-9  # m for r = 100 nm
w_res = ω_p/np.sqrt(ε_m_inf + 2*ε_h)

list = [5e-9, 10e-9, 20e-9, 40e-9, 100e-9]

for i in range (0, 5):

    r = list[i]

    γ = γ + (0.75*v_f/r)

    # Permittivity ε_m(ω) (complex value, not real)
    def ε_m(ω):
        return ε_m_inf - (ω_p**2) / (ω**2 + γ**2) - ( 1j * (ω_p**2 * γ/(ω**3 + γ**2 * ω) ) )

    # Clausius-Mossotti relation (Equation 15)
    def αj(ε_m):
        return 3.0 * v * (ε_m - 1) / (ε_m + 2)

    # Define the volume based on the radius r
    v = 4/3 * np.pi * r**3

    # Wavelength range from the graph
    wavelength_nm = np.linspace(460, 560, 1000)
    ω = 2.0 * np.pi * 3.0e8 / (wavelength_nm * 1.0e-9)  # Convert wavelength to angular frequency

    # Calculate alpha_j for the given range of ω
    α = αj(ε_m(ω))

    # Plotting the graph
    #plt.figure(figsize=(12, 6))
    plt.plot(wavelength_nm, np.abs(α/v), color=colors[i])

# Adding labels and title
plt.xlabel('Wavelength [nm]')
plt.ylabel('|α|/v [a.u.]')
plt.title('Corrections to quasistatic approximation')
plt.legend()

red_patch = mpatches.Patch(color='red', label='r = 100nm')
blue_patch = mpatches.Patch(color='blue', label='r = 40nm')
green_patch = mpatches.Patch(color='green', label='r = 20nm')
cyan_patch = mpatches.Patch(color='cyan', label='r = 10nm')
purple_patch = mpatches.Patch(color='purple', label='r = 5nm')
plt.legend(handles=[red_patch, blue_patch, green_patch, cyan_patch, purple_patch])

# Display the plot
plt.show()
```

## B. Modified long wave approximation (MLWA)
Previously, we analysed the resonance using dipolar approximation, which is valid as long as the size of nanoparticle $(d)$ is much smaller than the wavelength of the incident light $(\lambda)$. In this section, we will consider the finite size corrections due to the variation of incoming field, and hence the motion of the atoms in the nanoparticle. We do this by treating each atom in the nanoparticle as a dipole, and calculate the field due to it, including the retarding effects. Instead of setting $\textbf{k}·\textbf{r}$ to $0$, we consider Taylor series upto third order. This saves all the troubles one needs to go through while using brute-force computational method for Maxwell's equation's solution, and captures the correction to the simpler dipolar treatment. \\
Since the incoming light has oscillating electric field, the atoms will undergo harmonic motion leading the dipole moment given by:

$p=p_0 e^{i\omega t}e^{-ikr}, \quad ṗ=i\omega p, \quad p̈=-\omega^2 p\tag{41}$
*Calculation part* \\

Hence,

$E_R=\frac{2\cos{\theta}}{4\pi\epsilon_0\epsilon_h}\left(\frac{p}{r^3}+\frac{ṗ}{cr^2}\right), \quad E_\theta=\frac{\sin{\theta}}{4\pi\epsilon_0\epsilon_h}\left(\frac{p}{r^3}+\frac{ṗ}{cr^2}+\frac{p̈}{c^2 r}\right)\tag{42}$

Due to the polarisation density $\textbf{P}$ due to the incident field, electric field due to dipoles in a volume $dV$ is given by:
$dE_{||}=dE_R\cos{\theta}-dE_\theta\sin{\theta}, \quad dE_\perp=dE_R\sin{\theta}+dE_\theta\cos{\theta}\tag{43}$
Consider the perpendicular component first:

\begin{align}
E_\perp&=\int \left(\frac{2\cos{\theta}}{4\pi\epsilon_0\epsilon_h}\left(\frac{1}{r^3}+\frac{i\omega}{cr^2}\right)\sin{\theta} + \frac{\sin{\theta}}{4\pi\epsilon_0\epsilon_h}\left(\frac{1}{r^3}+\frac{i\omega}{cr^2}+\frac{-\omega^2}{c^2 r}\right)\cos{\theta}\right)Pr^2\sin{\theta}drd\theta d\phi \\
&=0
\tag{44}
\end{align}

since $\int_{0}^{2\pi}\cos{\theta}\sin^2{\theta}d\theta=0$. As for the parallel component:

\begin{align}
E_{||}&=\int \left(\frac{2\cos^2{\theta}}{4\pi\epsilon_0\epsilon_h}\left(\frac{1}{r^3}+\frac{i\omega}{cr^2}\right) - \frac{\sin^2{\theta}}{4\pi\epsilon_0\epsilon_h}\left(\frac{1}{r^3}+\frac{i\omega}{cr^2}+\frac{-\omega^2}{c^2 r}\right)\right)Pr^2\sin{\theta}drd\theta d\phi \\
&=\frac{P}{4\pi\epsilon_0\epsilon_h}\int\left[\frac{1}{r^3}(3\cos^2{\theta}-1)+\frac{k^2}{2r}(\cos^2{\theta}+1)-\frac{2ik^3}{3}\right]r^2\sin{\theta}d\theta d\phi \\
&=\frac{1}{4\pi\epsilon_0\epsilon_h}\frac{4\pi}{3}(k^2R^2-i\frac{2}{3}k^3R^3)P \\
\therefore E_{MLWA}&=\frac{1}{4\pi\epsilon_0\epsilon_h}(\frac{k^2}{R}-i\frac{2}{3}k^3)p_{np}
\tag{45}
\end{align}

where we have substituted $\omega=kc$ and volume integral is over a sphere of radius $R$. The formula can be generalized for ellipsoidal nanoparticles of semiaxies $a_j,j=x,y,z$:
$E_{MLWA,j}=\frac{1}{4\pi\epsilon_0\epsilon_h}(\frac{k^2}{a_j}-i\frac{2}{3}k^3)p_{np,j}\tag{46}$

From which we can write
$\alpha_{MLWA},j=\frac{\alpha_j}{1-\frac{\alpha_j}{4\pi}\left(\frac{k^2}{a_j}-\frac{2}{3}k^3\right)}\tag{47}$ \\

It is easily seen that for small nanoparticles, i.e. $k\rightarrow0$, we get the QSA approximation $(\alpha_{MLWA,j}\approx\alpha_j)$. The term $\alpha_j(k^2/a_j)$ obtained in dynamical calculations ($k>0$) is called dynamic depolarization. As its coefficient is real, it corresponds to the change in effective particle depolarization factor. As the particle size increases, more negative value of $\epsilon_1$ is necessary to meet resonance condition. \\
Whereas the imaginary term $i\alpha_j\frac{2}{3}k^3$ comes due to  spontaneous emission of radiation by the induced dipole, and hence is called radiative damping. This terms contributes to $\epsilon_2$ and hence to resonance damping. For larger particle sizes, this causes a broadening and decrease in plasmonic resonance

## Derivation of Electric field due to Oscillating dipole
We use a new pair of potential functions, namely $\textbf{\Pi}_e$ and $\textbf{\Pi}_m$. These are called \textit{Hertz vectors} or \textit{Polarisation potentials}. They relate to usual potentials as:

\begin{align}
\bar{A}&=\frac{1}{c}\bar{\Pi}_e \\
\phi&=-\nabla·\bar{\Pi}_e
\tag{48}
\end{align}

And hence we can express the fields as:
$\bar{E}=\nabla\times\left(-\frac{1}{c}\dot{\bar{\Pi}}_m+\nabla\times\bar{\Pi}_e\right)+\left(\nabla^2\bar{\Pi}_e-\frac{1}{c^2}\ddot{\bar{\Pi}}_e\right)\tag{49}$
$\bar{B}=\nabla\times\left(\frac{1}{c}\dot{\bar{\Pi}}_e+\nabla\times\bar{\Pi}_m\right)\tag{50}$

Note that the Lorenz condition is satisfied: $\nabla·\bar{A}+\frac{1}{c}\frac{\partial}{\partial t}\phi=0$. Then the Maxwell's equations can be rewritten as:
\begin{align}
\nabla^2\bar{\Pi}_e-\frac{1}{c^2}\ddot{\bar{\Pi}}_e&=-4\pi \bar{P}
\tag{51} \\
\nabla^2\bar{\Pi}_m-\frac{1}{c^2}\ddot{\bar{\Pi}}_m&=-4\pi \bar{M}
\tag{52}
\end{align}
where $P$ and $M$ are the polarization and magnetization respectively. We can express these in the form of retarded potential:
\begin{align}
\bar{\Pi}_e&=\int\frac{\bar{[P]}}{R}dV' \tag{53} \\
\bar{\Pi}_m&=\int\frac{\bar{[M]}}{R}dV' \tag{54}
\end{align}

Now we consider a dipole vibrating in a fixed direction $\hat{n}$. Such a dipole moment is characterized by electric polarizarion:
$\bar{P}(\bar{r} ,t)=p(t)\delta(\bar{r}-\bar{r}_0)\hat{n}\tag{55}$
The Hertz vector is given by:
$\bar{\Pi}_e=\frac{p(t-\frac{R}{c})}{R}\hat{n}, \quad \bar{\Pi}_m=0\tag{56}$
Substituting this in the expression for $\bar{E}$,
$\bar{E}=\nabla\times\nabla\times\bar{\Pi}_e\tag{57}$
Using the identity $\nabla\times\nabla\times A\equiv\nabla(\nabla·\bar{A})-\nabla^2\bar{A}$ and $\nabla^2\bar{\Pi}_e=\frac{1}{c^2}\ddot{\bar{\Pi}}_e$, we get:
$\bar{E}=\nabla(\nabla·\bar{\Pi}_e)-\frac{1}{c^2}\ddot{\bar{\Pi}}_e\tag{58}$
Substituting the expression of $\bar{\Pi}_e$, we get
$\bar{E}=\left(\frac{3[p]}{R^5}+\frac{3[\dot{p}]}{cR^4}+\frac{[\ddot{p}]}{c^2R^3}\right)(\hat{n}·\bar{R})\bar{R}-\left(\frac{[p]}{R^3}+\frac{[\dot{p}]}{cR^2}+\frac{[\ddot{p}]}{c^2R}\right)\hat{n}\tag{59}$
To further simplify this set the $z-$axis along $\hat{n}$ so that $\bar{R}=R\hat{r}$ and $\hat{n}=\cos{\theta}\hat{r}-\sin{\theta}\hat{\theta}$ and hence $\bar{R}·\hat{n}=R\cos{\theta}$
hence the radial and angular components of the electric field are given by
\begin{align}
E_R&=2\left(\frac{[p]}{R^3}+\frac{[\dot{p}]}{cR^2}\right)\cos{\theta} \tag{60} \\
E_\theta&=\left(\frac{[p]}{R^3}+\frac{[\dot{p}]}{cR^2}+\frac{[\ddot{p}]}{c^2R}\right)\sin{\theta} \tag{61}
\end{align}
